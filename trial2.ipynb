```python
import pandas as pd
```

```python
movies_data = 'ml-25m/movies.csv'
ratings_data = 'ml-25m/ratings.csv'
```

```python
pd.set_option('display.max_rows', 20)
```

```python
missing_values = ['na','--','?','-','None','none','non']
```

```python
movies_df = pd.read_csv(movies_data, na_values=missing_values)
ratings_df = pd.read_csv(ratings_data, na_values=missing_values)
```

```python
print('Movies_df Shape:',movies_df.shape)
movies_df
```

```python
movies_df['year'] = movies_df.title.str.extract('(\(\d\d\d\d\))',expand=False)
#Removing the parentheses.
movies_df['year'] = movies_df.year.str.extract('(\d\d\d\d)',expand=False)
#Removing the years from the 'title' column.
movies_df['title'] = movies_df['title'].str.replace('(\(\d\d\d\d\))', '', regex=True)
#Applying the strip function to get rid of any ending white space characters that may have appeared, using lambda function.
movies_df['title'] = movies_df['title'].apply(lambda x: x.strip())
movies_df
```

```python
movies_df['genres'] = movies_df.genres.str.split('|')
movies_df.head()
```

```python
movies_df.dtypes
```

```python
# Filling year non num values with zeros.
movies_df.year.fillna(0, inplace=True)
# Converting columns year from obj to int16 and movieId from int64 to int32 to save memory.
movies_df.year = movies_df.year.astype('int16')
movies_df.movieId = movies_df.movieId.astype('int32')
```

```python
movies_df.dtypes
```

```python
#make a copy of the movies_df.
movies_with_genres = movies_df.copy(deep=True)
#iterate through movies_df, then append the movie genres as columns of 1s or 0s.
#1 if that column contains movies in the genre at the present index and 0 if not.
x = []
for index, row in movies_df.iterrows():
    x.append(index)
    for genre in row['genres']:
        movies_with_genres.at[index, genre] = 1
# Confirm that every row has been iterated and acted upon.
print(len(x) == len(movies_df))
movies_with_genres.head(3)
```

```python
#Filling in the non num values with 0 to show that a movie doesn't have that column's genre.
movies_with_genres = movies_with_genres.fillna(0)
movies_with_genres.head(3)
```

```python
# print out the shape and first five rows of ratings data.
print('Ratings_df shape:',ratings_df.shape)         
ratings_df.head()
```

```python
#drop the timestamp column
ratings_df.drop('timestamp', axis=1, inplace=True)
#confirm the drop
ratings_df.head(3)

ratings_df.dtypes
```

```python
#someones list
person_movie_ratings = [
            {'title':'Predator', 'rating':4.9},
            {'title':'Final Destination', 'rating':4.9},
            {'title':'Mission Impossible', 'rating':4},
            {'title':"Beverly Hills Cop", 'rating':3},
            {'title':'Exorcist, The', 'rating':4.8},
            {'title':'Waiting to Exhale', 'rating':3.9},
            {'title':'Avengers, The', 'rating':4.5},
            {'title':'Omen, The', 'rating':5.0}
         ]
person_movie_ratings = pd.DataFrame(person_movie_ratings)
person_movie_ratings
```

```python
#Extract movie Ids from movies_df and update person_movie_ratings with movie Ids
person_movie_Id = movies_df[movies_df['title'].isin(person_movie_ratings['title'])]
#merge person movie Id and ratings into the person_movie_ratings data frame. 
person_movie_ratings = pd.merge(person_movie_Id, person_movie_ratings)
person_movie_ratings
```

```python
# Check if the columns 'genres' and 'year' exist in 'person_movie_ratings' before dropping them
if 'genres' in person_movie_ratings.columns and 'year' in person_movie_ratings.columns:
    # Dropping 'genres' and 'year' columns from 'person_movie_ratings'
    person_movie_ratings = person_movie_ratings.drop(['genres', 'year'], axis=1)
    print(person_movie_ratings)
    
    # Filter the selection by outputting movies that exist in both 'person_movie_ratings' and 'movies_with_genres'
    person_genres_df = movies_with_genres[movies_with_genres.movieId.isin(person_movie_ratings.movieId)]
    print(person_genres_df)
else:
    print("Error: 'genres' and/or 'year' columns not found in 'person_movie_ratings'")
```

```python

#reset index to default and drop the existing index.
person_genres_df.reset_index(drop=True, inplace=True)

#drop redundant columns using .loc
person_genres_df = person_genres_df.loc[:, ~person_genres_df.columns.isin(['movieId', 'title', 'genres', 'year'])]

person_genres_df

print('Shape of person_movie_ratings is:',person_movie_ratings.shape)
print('Shape of person_genres_df is:',person_genres_df.shape)
```

```python
#find the dot product of transpose of person_genres_df by person rating column.
person_profile = person_genres_df.T.dot(person_movie_ratings.rating)
person_profile
```

```python
print(person_profile.sum())
```

```python
#set the index to the movieId.
movies_with_genres = movies_with_genres.set_index(movies_with_genres.movieId)
movies_with_genres.head()
#del four unnecessary columns.
movies_with_genres.drop(['movieId','title','genres','year'], axis=1, inplace=True)
movies_with_genres.head()
```

```python
#multiply the genres by the weights and then take the weighted average
recommendation_table_df = (movies_with_genres.dot(person_profile)) / person_profile.sum()
```

```python
#sort values from highest to lowesttt
recommendation_table_df.sort_values(ascending=False, inplace=True)
recommendation_table_df.head()
```

```python
#make a copy of the original movies_df
copy = movies_df.copy(deep=True)
#set its index to movieId, ie make index values as movieid 
copy = copy.set_index('movieId', drop=True)
#list out the top 20 recommended movieIds
top_20_index = recommendation_table_df.index[:20].tolist()
#slice these indices from the copied movies df and save in a variable
recommended_movies = copy.loc[top_20_index, :]
#display the top 20 movies in descending order of preference
recommended_movies
```
